3. What are pre-defined keywords in Python
  - Pre-defined keywords in Python are reserved words that have special meaning in the language. we cannot use them as variable names, function names, or any other identifiers, because they define the structure and behavior of Python programs
  - Control flow (if, else, for, while)
- Function and class definitions (def, class, return)
- Logical operations (and, or, not)
- Exception handling (try, except, finally)
- Scope and namespace (global, nonlocal)
- Asynchronous programming (async, await)



4. Can keywords be used as variable names
  - No, Python keywords cannot be used as variable names

```python
def test_args(a_list, an_int):
    a_list.append(100)
    an_int += 10
    print("List inside:", a_list)
    print("Int inside:", an_int)

lst = [1, 2, 3]
num = 10
test_args(lst, num)
print("List outside:", lst)
print("Int outside:", num)
```

28. Reverse a string using while loop:

```python
s = input("Enter text: ")
rev = ""
i = len(s) - 1
while i >= 0:
    rev += s[i]
    i -= 1
print(rev)
```

29. Factorial of a number using while loop:

```python
n = int(input("Number: "))
result = 1
i = 1
while i <= n:
    result *= i
    i += 1
print("Factorial:", result)
```
